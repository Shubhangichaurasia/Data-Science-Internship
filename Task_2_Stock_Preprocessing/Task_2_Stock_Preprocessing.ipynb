```python
# 1. Convert the 'date' column to a real datetime format
# (This fixes the "Object" data type issue)
df['date'] = pd.to_datetime(df['date'])

# 2. Filter the data to keep only Apple (AAPL)
# (We use .copy() to tell Python this is a new, separate table, not just a view of the old one)
df_clean = df[df['symbol'] == 'AAPL'].copy()

# 3. Sort chronologically from Oldest to Newest
# (Crucial for time-series. If this is wrong, 'Forward Fill' will fill from the future!)
df_clean = df_clean.sort_values('date')

# 4. Set the Date as the Index
# (Now the rows will be labeled by time, not 0, 1, 2...)
df_clean.set_index('date', inplace=True)

# 5. Check the result
print("--- Data Shape (Rows, Columns) ---")
print(df_clean.shape)
print("\n--- First 5 Rows ---")
df_clean.head()
```

```python
# 1. Fill missing values using Forward Fill ('ffill')
# This propagates the last valid observation forward to next valid
df_clean = df_clean.fillna(method='ffill')

# 2. Double check: Are there any missing values left?
# (This should print all Zeros now)
print("--- Missing Values After Cleaning ---")
print(df_clean.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler

# 1. Initialize the Scaler (scales data between 0 and 1)
scaler = MinMaxScaler()

# 2. Apply the scaler to our clean data
# (Note: This returns a numpy array, so we convert it back to a DataFrame to keep column names)
df_scaled = pd.DataFrame(scaler.fit_transform(df_clean), 
                         columns=df_clean.columns, 
                         index=df_clean.index)

# 3. Split into Train (First 80%) and Test (Last 20%)
# We use .iloc to slice strictly by order, ensuring no shuffling happens
train_size = int(len(df_scaled) * 0.8)

train_data = df_scaled.iloc[:train_size]
test_data = df_scaled.iloc[train_size:]

# 4. Verify the split
print(f"Total Days: {len(df_scaled)}")
print(f"Training Days (First 80%): {len(train_data)}")
print(f"Testing Days (Last 20%):   {len(test_data)}")

# Show the scaled data (values should be between 0.0 and 1.0)
df_scaled.head()
```

```python
# 1. Drop the 'symbol' column (It's text, and we don't need it anymore)
# (We use errors='ignore' so it won't crash if you already dropped it)
df_clean = df_clean.drop('symbol', axis=1, errors='ignore')

# 2. Check to make sure only numbers are left
print("Columns left:", df_clean.columns.tolist())

# 3. NOW run the Scaling code again
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_clean), 
                         columns=df_clean.columns, 
                         index=df_clean.index)

# 4. Split into Train/Test again
train_size = int(len(df_scaled) * 0.8)
train_data = df_scaled.iloc[:train_size]
test_data = df_scaled.iloc[train_size:]

print("Success! Data Scaled and Split.")
df_scaled.head()
```

```python
# Save the processed data to a new CSV file
# We include 'index=True' to keep the Dates!
df_scaled.to_csv('processed_aapl_data.csv')

print("Task 2 Completed! File saved as 'processed_aapl_data.csv'")
```
